````python
# =============================================================================
# Cell 1 — Config
# CLEVRTex Dataset Download
# Run this notebook ONCE on Kaggle (Internet must be enabled) to download
# CLEVRTex and persist it as a Kaggle Dataset the training notebook can attach.
# Expected runtime: ~30–60 minutes.
# =============================================================================

import os
import subprocess
import shutil
from pathlib import Path

OUT_DIR = "/kaggle/working"  # clevrtex_full/ subfolder will be created here

PARTS = [
    "https://thor.robots.ox.ac.uk/datasets/clevrtex/clevrtex_full_part1.tar.gz",
    "https://thor.robots.ox.ac.uk/datasets/clevrtex/clevrtex_full_part2.tar.gz",
    "https://thor.robots.ox.ac.uk/datasets/clevrtex/clevrtex_full_part3.tar.gz",
    "https://thor.robots.ox.ac.uk/datasets/clevrtex/clevrtex_full_part4.tar.gz",
    "https://thor.robots.ox.ac.uk/datasets/clevrtex/clevrtex_full_part5.tar.gz",
]
````

````python
# =============================================================================
# Cell 2 — Download and extract (one part at a time to stay within 20 GB limit)
# =============================================================================

for url in PARTS:
    filename = url.split("/")[-1]
    tarball = os.path.join(OUT_DIR, filename)

    used_gb = shutil.disk_usage(OUT_DIR).used / 1e9
    print(f"\n--- Downloading {filename} (disk used before: {used_gb:.2f} GB) ---")

    subprocess.run(["wget", "-q", "--show-progress", url, "-P", OUT_DIR], check=True)

    print(f"Extracting {filename} ...")
    subprocess.run(["tar", "-xzf", tarball, "-C", OUT_DIR], check=True)

    os.remove(tarball)

    used_gb = shutil.disk_usage(OUT_DIR).used / 1e9
    print(f"Done with {filename}. Disk used after removal: {used_gb:.2f} GB")

print("\nAll parts downloaded and extracted.")
````

````python
# =============================================================================
# Cell 3 — Verification
# =============================================================================

clevrtex_dir = Path("/kaggle/working/clevrtex_full")

images = list(clevrtex_dir.rglob("CLEVRTEX_full_??????.png"))
print(f"Image count:  {len(images)}")
assert len(images) > 0, "No images found — check that extraction succeeded."

masks = list(clevrtex_dir.rglob("*_flat.png"))
print(f"Mask count:   {len(masks)}")

total_gb = sum(f.stat().st_size for f in clevrtex_dir.rglob("*") if f.is_file()) / 1e9
print(f"Total size:   {total_gb:.2f} GB")

print()
print("\u2713 Dataset ready. Attach this notebook's output to your training notebook as an input dataset.")
print("  Set DATA_ROOT = '/kaggle/input/<this-notebook-slug>' in Cell 1 of the training notebook.")
````
